```python
import rootutils
from pathlib import Path
PROJECT_ROOT = rootutils.setup_root(
    search_from=Path.cwd(), dotenv=True, pythonpath=True, cwd=False)

import scipy.io as io
import polars as pl
from pathlib import Path
```

### Convert .mat hyperparams file to csv

```python
filename = PROJECT_ROOT / "data/hyperparameters/hyperparameters.mat"
file = io.matlab.loadmat(filename)

type(file)
# for key, val in file.items():
#     print(key, val)

print(file["hyperparam_hist"].shape)

outputs = ["yaw", "pos_0", "pos_1", "pos_2"]

for i, output in enumerate(outputs) :
    hyperparams = file["hyperparam_hist"][:,:,i].T
    print(hyperparams.shape)
    hyperparams[1,:] = hyperparams[1,:]
    
    df = pl.from_numpy(hyperparams, ["LogLikelihood", "length_scale", "sigma_f", "sigma_n"])
    csv_filename = PROJECT_ROOT / f"data/hyperparameters/matlab/hyperparams_{output}.csv"
    df.write_csv(csv_filename)
    
```
